```octave
% 2D Sinusoid Plot
t = 0:0.01:2*pi;
y = sin(t);

figure;
subplot(2, 1, 1); % Create a subplot with 2 rows, 1 column, and plot in the 1st part
plot(t, y);
title('Sine Wave');
xlabel('Time (s)');
ylabel('Amplitude');
grid on;

% 3D Plot
[x, y] = meshgrid(-2:0.1:2, -2:0.1:2);
z = sin(pi * x) .* cos(pi * y);

subplot(2, 1, 2); % Plot in the 2nd part
mesh(x, y, z);
title('3D Mesh Plot of sin(pi * x) * cos(pi * y)');
xlabel('X');
ylabel('Y');
zlabel('Z');
grid on;
```

```octave
% Ensure gnuplot is the active toolkit
graphics_toolkit('gnuplot');

% 2D Sinusoid Plot
t = 0:0.01:2*pi;
y = sin(t);

figure;
subplot(2, 1, 1); % Create a subplot with 2 rows, 1 column, and plot in the 1st part
plot(t, y);
title('Sine Wave');
xlabel('Time (s)');
ylabel('Amplitude');
grid on;

% 3D Plot
[x, y] = meshgrid(-2:0.1:2, -2:0.1:2);
z = sin(pi * x) .* cos(pi * y);

subplot(2, 1, 2); % Plot in the 2nd part
mesh(x, y, z);
title('3D Mesh Plot of sin(pi * x) * cos(pi * y)');
xlabel('X');
ylabel('Y');
zlabel('Z');
grid on;
```

```octave
% Model Parameters
graphics_toolkit('gnuplot');
m = 60; L = 2; g=9.81; mu = 2e2;

a = g/L; b = mu/(m*L^2); c = 1/(m*L^2)

% Time Interval
T = [0 15];

% Initial Conditions
x0 = [pi/2; 0];

% Constant Input
u = 0;

% System Model
sys = @(t,x) [ x(2)
              -a*sin(x(1))-b*x(2)+c*u];

% Solve the System
[t,x] = ode45(sys,T,x0);

% Plot State Signal
subplot(2,1,1)
plot(t,x(:,1)); grid on
xlabel('t'); ylabel('x_1')
title('Pendulum States')

subplot(2,1,2)
plot(t,x(:,2)); grid on
xlabel('t'); ylabel('x_2')

% Time-varying input
u = @(t) 2e2*sin(sqrt(a)*t);

% System Model
sys = @(t,x) [ x(2)
              -a*sin(x(1))-b*x(2)+c*u(t)];

% Solve the System
[t,x] = ode45(sys,T,x0);

% Plot State Signal
figure
subplot(2,1,1)
plot(t,x(:,1)); grid on
xlabel('t'); ylabel('x_1')
title('Pendulum States')

subplot(2,1,2)
plot(t,x(:,2)); grid on
xlabel('t'); ylabel('x_2')

% Plot Input Signal
figure
plot(t,u(t)); grid on
xlabel('t'); ylabel('u')
title('Pendulum Input')

% Dynamic Model
f = @(x,u) [ x(2)
            -a*sin(x(1))-b*x(2)+c*u];

% Feedback Controller
u = @(x) a/c*sin(pi/4)-1e3*(x(1)-pi/4)-5e2*x(2);

% System Model
sys = @(t,x) f(x,u(x));

% Solve the System
[t,x] = ode45(sys,T,x0);

% Plot State Signal
figure
subplot(2,1,1)
plot(t,x(:,1)); grid on
xlabel('t'); ylabel('x_1')
title('Pendulum States')

subplot(2,1,2)
plot(t,x(:,2)); grid on
xlabel('t'); ylabel('x_2')

% Evaluate u(x) at every time instant
u_t = zeros(length(t),1);
for i = 1:length(t)
    u_t(i)=u(x(i,:)');
end

% Plot the Input Signal
figure
plot(t,u_t); grid on
xlabel('t'); ylabel('u')
title('Pendulum Input')
```
